```python
corpus_raw = u""
for book_filename in book_filenames:
    print("Reading '{0}'...".format(book_filename))
    with codecs.open(book_filename, "r", "utf-8") as book_file:
        corpus_raw += book_file.read()
    print("Corpus is now {0} characters long".format(len(corpus_raw)))
    print()
```

```python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
raw_sentences = tokenizer.tokenize(corpus_raw)

print(raw_sentences[3])
```

```python
def cleanText(corpus):
    punctuation = """.,?!:;(){}[]"""
    corpus = [z.lower().replace('\n','') for z in corpus]
    corpus = [z.replace('<br />', ' ') for z in corpus]
#treat punctuation as individual words
    for c in punctuation:
        corpus = [z.replace(c, ' %s '%c) for z in corpus]
    corpus = [z.split() for z in corpus]
    return corpus
def tagglizeDoc(data, tag):
    tagglized = []
    for i,v in enumerate(data):
        t = tag[i]
        tagglized.append(LabeledSentence(v, [t]))
    return tagglized
print ("reading file")
path = "icse.txt"
data=[]
label=[]

for line in open(path,"r",encoding="utf8"):
    temp=line.split("\t");
    label.append(temp[0])
    data.append(temp[1])
    print (temp[0],temp[1])
    
#do some simple data cleanig
data = cleanText(data)
data = tagglizeDoc(data,label)
print(data)
corpus = list(data)
```

```python
import gensim.models as g
import logging
import sys
import tokenizer
from multiprocessing import cpu_count

model = g.Doc2Vec(vector_size=50, min_count=2, epochs=35)

#save model
model.build_vocab(corpus)
model.train(corpus, total_examples=model.corpus_count, epochs=model.epochs)
model.save('laws_model.bin')
print(model)
```
